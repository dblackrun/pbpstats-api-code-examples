This endpoint can be used to get each team's relative offensive and defensive efficiency relative to the league average

```python
import requests
```

### NBA

```python
url = "https://api.pbpstats.com/get-relative-off-def-efficiency/nba"
response = requests.get(url)
response_json = response.json()
response_json['results']
```

### WNBA

```python
url = "https://api.pbpstats.com/get-relative-off-def-efficiency/wnba"
response = requests.get(url)
response_json = response.json()
response_json['results']
```
