This endpoint is used to get top results for a season or game

```python
import requests
```

### Most corner 3 makes by a team in a season

```python
url = "https://api.pbpstats.com/get-top-results/nba"
params = {
    "SeasonType": "Regular Season",
    "EntityType": "Team", # Options are Team, Opponent, Player, Lineup, LineupOpponent
    "Stat": "Corner3FGM",
    "Type": "season"
}
response = requests.get(url, params=params)
response_json = response.json()
response_json["results"]
```

### Fewest corner 3 makes by a team in a season

```python
url = "https://api.pbpstats.com/get-top-results/nba"
params = {
    "SeasonType": "Regular Season",
    "EntityType": "Team", # Options are Team, Opponent, Player, Lineup, LineupOpponent
    "Stat": "Corner3FGM",
    "Type": "season",
    "SortOrder": "Asc"
}
response = requests.get(url, params=params)
response_json = response.json()
response_json["results"]
```
