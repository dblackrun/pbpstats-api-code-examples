This endpoint can be used to get a breakdown of when a player was on the floor for each game

```python
import requests
```

### Kevin Durant

```python
url = "https://api.pbpstats.com/get-playing-time-distribution/nba"
params = {
    "TeamId": "1610612751",
    "Season": "2021-22",
    "SeasonType": "Regular Season",
    "PlayerId": "201142"
}
response = requests.get(url, params=params)
response_json = response.json()
response_json["results"] # start and end time in possession item are seconds into the game at the start and end of the possession
```
