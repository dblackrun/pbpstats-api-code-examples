This endpoint can be used to get game stats by player/lineup

```python
import requests
```

### NBA - player

```python
url = "https://api.pbpstats.com/get-game-stats"
params = {
    "GameId": "0022100001",
    "Type": "Player" # options are Player, Lineup and LineupOpponent
}
response = requests.get(url, params=params)
response_json = response.json()
response_json['stats'] # dictionary with home and away stats by period
```

### WNBA - lineup

```python
url = "https://api.pbpstats.com/get-game-stats"
params = {
    "GameId": "1022100001",
    "Type": "Lineup" # options are Player, Lineup and LineupOpponent
}
response = requests.get(url, params=params)
response_json = response.json()
response_json['stats']
```
