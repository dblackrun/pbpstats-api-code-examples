This endpoint can be used to get assist network data

```python
import requests
```

### Team for season

```python
url = "https://api.pbpstats.com/get-assist-networks/nba"
params = {
    "Season": "2021-22",
    "SeasonType": "Regular Season",
    "EntityType": "Team", # Options: Team, Lineup
    "EntityId": "1610612737"
}
response = requests.get(url, params=params)
response_json = response.json()
response_json["results"]["links"] # source is assist from player id, target is assist to player id
```

### Lineup for season

```python
url = "https://api.pbpstats.com/get-assist-networks/nba"
params = {
    "Season": "2021-22",
    "SeasonType": "Regular Season",
    "EntityType": "Lineup", # Options: Team, Lineup
    "EntityId": "1628381-1629027-1629631-203991-203992"
}
response = requests.get(url, params=params)
response_json = response.json()
response_json["results"]["links"] # source is assist from player id, target is assist to player id
```
