This endpoint can be used to get live game stats. Note that since this deals with live data, there can be issues with the play-by-play data that get fixed eventually. I'm impressed with the quality of the data given it is in real time.

```python
import requests
```

### Team Stats

```python
game_id = "1022200047"
result_type = "team"
url = f"https://api.pbpstats.com/live/game/{game_id}/{result_type}"
response = requests.get(url)
response_json = response.json()
response_json
```
