This endpoint can be used to get on/off stats for query parameters

```python
import requests
```

### Player stats - Stephen Curry on/off

```python
url = "https://api.pbpstats.com/get-on-off/nba/player"
params = {
    "Season": "2021-22",
    "SeasonType": "Regular Season",
    "TeamId": "1610612744",
    "PlayerId": "201939"
}
response = requests.get(url, params=params)
response_json = response.json()
response_json['results'] # this is a dict with stats for each teammate by stat
```
