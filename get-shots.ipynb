This endpoint can be used to get shots based on score and time parameters

```python
import requests
```

### Blocked 3s by the Raptors

```python
url = "https://api.pbpstats.com/get-shots/nba"
params = {
    "Season": "2021-22",
    "SeasonType": "Regular Season",
    "EntityType": "Opponent",
    "EntityId": "1610612761",
    "StartType": "All",
    "Blocked": "true",
    "ShotValue": 3
}
response = requests.get(url, params=params)
response_json = response.json()
response_json["results"]
```

### Hawks shots within 4 seconds of getting an offensive rebound from a missed 3

```python
url = "https://api.pbpstats.com/get-shots/nba"
params = {
    "Season": "2021-22",
    "SeasonType": "Regular Season",
    "EntityType": "Opponent",
    "EntityId": "1610612737",
    "StartType": "All",
    "Blocked": "true",
    "OrebShotType": "Missed3", # set possible values for OrebShotType below
    "SecondsSinceOrebLte": 4
}
response = requests.get(url, params=params)
response_json = response.json()
response_json["results"]
```

#### possible values for OrebShotType

```python
oreb_shot_types = [
    "MissedFG", # Any Missed FG
    "Missed2", # Any Missed 2
    "MissedMidRange", # Any Mid Range Miss
    "Missed3", # Any Missed 3
    "AtRim", # Any At Rim Miss
    "AtRimBlocked", # Any At Rim Blocked
    "ShortMidRange", # Any Short Mid Range Miss
    "LongMidRange", # Any Long Mid Range Miss
    "Arc3", # Any Above the Break 3 Miss
    "Corner3", # Any Corner 3 Miss
    "FT", # Any Free Throw Miss
]
```
