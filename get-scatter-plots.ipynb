This endpoint can be used to get stats for each team for the selected stats

```python
import requests
```

### Points per 100 possessions and seconds per possession

```python
url = "https://api.pbpstats.com/get-scatter-plots/nba"
params = {
    "Season": "2021-22",
    "SeasonType": "Regular Season",
    "Xaxis": "PtsPer100Poss", # see list of available stat keys below
    "Yaxis": "SecondsPerPoss", # see list of available stat keys below
    "XaxisType": "Team", # Team or Opponent
    "YaxisType": "Team" # Team or Opponent
}
response = requests.get(url, params=params)
response_json = response.json()
response_json['results']
```
