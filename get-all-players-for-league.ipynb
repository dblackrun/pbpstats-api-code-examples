This endpoint gives us a dictionary mapping the player id to the player name for all players that are in the database for the league

```python
import requests
```

## NBA

```python
players_response = requests.get("https://api.pbpstats.com/get-all-players-for-league/nba")
players = players_response.json()
players["players"]
```

## WNBA

```python
players_response = requests.get("https://api.pbpstats.com/get-all-players-for-league/wnba")
players = players_response.json()
players["players"]
```
