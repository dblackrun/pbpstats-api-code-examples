This endpoint can be used to get a breakdown of when each combo was on the floor for all on/off combinations of selected players for each game

```python
import requests
```

### Sixers for James Harden and Joel Embiid

```python
wowy_url = "https://api.pbpstats.com/get-wowy-combo-playing-time-distribution/nba"
wowy_params = {
    "TeamId": "1610612755", # Sixers
    "Season": "2021-22",
    "SeasonType": "Regular Season",
    "PlayerIds": "201935,203954", # comma separated player ids - Harden and Embiid
    "OnlyShowAllPlayersOn": "false" # When true, only returns results for when all players are on
}
wowy_response = requests.get(wowy_url, params=wowy_params)
wowy = wowy_response.json()
wowy["results"] # start and end time in possession item are seconds into the game at the start and end of the possession
```
