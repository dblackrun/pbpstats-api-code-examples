This endpoint can be used to get the frequency and net efficiency for each team by the number of starters on the floor. Starters are defined on a game-by-game basis

```python
import requests
```

### NBA

```python
url = "https://api.pbpstats.com/get-number-of-starter-stats-by-team/nba"
params = {
    "Season": "2021-22", # to query multiple seasons, separate seasons by a comma
    "SeasonType": "Regular Season"
}
response = requests.get(url, params=params)
response_json = response.json()
response_json['results']
```
